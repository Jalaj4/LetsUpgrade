```python
import pandas as pd
```

```python
dataset = pd.read_csv("general_data.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset.Attrition = le.fit_transform(dataset.Attrition)
dataset.BusinessTravel  = le.fit_transform(dataset.BusinessTravel)
dataset.Department  = le.fit_transform(dataset.Department)
dataset.EducationField  = le.fit_transform(dataset.EducationField)
dataset.JobRole  = le.fit_transform(dataset.JobRole)
dataset.Gender  = le.fit_transform(dataset.Gender)
dataset['Over18'] = le.fit_transform(dataset.Over18)
dataset.MaritalStatus  = le.fit_transform(dataset.MaritalStatus)
s=[]
r=[]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.fillna(method='bfill')
dataset = dataset.fillna(method='ffill')
dataset.isna().sum()
```

```python
y = dataset.Attrition
x = dataset[['Age','BusinessTravel','Department','DistanceFromHome','Education','EducationField','EmployeeCount','EmployeeID','Gender','JobLevel','JobRole','MaritalStatus','MonthlyIncome','PercentSalaryHike','StandardHours','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
import statsmodels.api as sm
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
```

```python
result = Logistic.fit()
result.summary()
```

1. Age,MaritalStatus,TrainingTimesLastYear,YearsSinceLastPromotion,YearsWithCurrManager are most significant for attrition
2. P value 0.86 is greater then 0.05 so the probibality of attrition is more
